```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# 입력 디렉토리 및 출력 디렉토리 설정
input_dir = "/content/drive/MyDrive/DSL/2023_2/모델링 프로젝트/Data/genres_original"
output_dir = "/content/drive/MyDrive/딥러닝과응용/spectrogram"

# 모든 장르 디렉토리에 대해 반복
for genre in os.listdir(input_dir):
    genre_dir = os.path.join(input_dir, genre)

    # 장르 디렉토리 내 모든 오디오 파일에 대해 반복
    for audio_file in os.listdir(genre_dir):
        audio_path = os.path.join(genre_dir, audio_file)

        try:
            # 오디오 파일 로드
            y, sr = librosa.load(audio_path)

            # 멜 스펙트로그램 생성
            mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
            mel_spect = librosa.power_to_db(mel_spect, ref=np.max)

            # 스펙트로그램 이미지 표시 및 저장
            librosa.display.specshow(mel_spect, fmax=8000)

            # 이미지 저장
            output_file = os.path.join(output_dir, genre, f"{audio_file[:-4]}.png")
            os.makedirs(os.path.dirname(output_file), exist_ok=True)
            plt.savefig(output_file, bbox_inches='tight')
            plt.close()

        except Exception as e:
            # Print the error (you can customize this part)
            print(f"Error processing {audio_path}: {e}")
            # Continue to the next iteration
            continue
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
import random as rn
from IPython.display import Image
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import os
%matplotlib inline
```

```python
dir_path = "/content/drive/MyDrive/딥러닝과응용/spectrogram" # Spectrogram 데이터 경로.

filenames = os.listdir(dir_path) # listdir은 폴더안에 있는 파일들의 이름을 불러오는 함수
filenames

categories = os.listdir(dir_path)
# 각 image의 종류
num_classes = len(categories)
# 전체 종류의 갯수 (9개)
print(num_classes)
```

```python
Y = []
X = []

for idex, category in enumerate(categories):
    # 각 category에 대해 index와 함께 순회하면서
    label = [0] * num_classes  # 클래스 개수만큼 0으로 초기화된 리스트 생성
    label[idex] = 1  # 해당 category에 대한 위치에 1을 넣어서 정답 레이블 생성
    image_dir = dir_path + '/' + category + '/'

    # 폴더 내 파일 리스트 가져오기
    file_list = os.listdir(image_dir)
    num_files = len(file_list)  # 파일 개수 계산
    print(f"{category}: {num_files} files found")  # 폴더명과 파일 개수 출력

    # 파일 리스트에 대해 반복하면서 이미지 읽기 및 처리
    for filename in file_list:
        if not filename.endswith('png'):
            continue
        url = os.path.join(image_dir, filename)
        img = cv2.imread(url)  # 이미지 읽기
        img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_NEAREST)  # 이미지 크기 조정

        X.append(img)  # 이미지 리스트에 추가
        Y.append(label)  # 정답 레이블 리스트에 추가
```

```python
X = np.array(X)
Y = np.array(Y)

print("Shape of X:", X.shape) # 샘플수, 높이, 너비, 채널
print("Shape of Y:", Y.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 클래스마다 첫 번째 이미지와 클래스 이름을 저장할 리스트 생성
class_first_images = []
# X와 Y 리스트에서 각 클래스에 해당하는 첫 번째 이미지와 클래스 이름을 class_first_images에 저장
for i in range(num_classes):
    images = [X[j] for j in range(len(X)) if np.argmax(Y[j]) == i]  # 해당 클래스에 해당하는 이미지 리스트
    first_image = images[0]  # 첫 번째 이미지 선택
    class_name = categories[i]
    class_first_images.append((first_image, class_name))  # class_first_images 리스트에 추가
# 각 클래스에서 첫 번째 이미지 선택해서 플롯으로 보여주기
fig, axes = plt.subplots(nrows=1, ncols=num_classes, figsize=(10, 4))
for i, (image, class_name) in enumerate(class_first_images):
    axes[i].imshow(image)  # 이미지 플롯에 표시
    axes[i].axis("off")
    axes[i].set_title(class_name)
plt.tight_layout()
plt.show()
```

# ResNet50

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Dense , Conv2D , Dropout , Flatten , Activation, MaxPooling2D , GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam , RMSprop
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau , EarlyStopping , ModelCheckpoint , LearningRateScheduler
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# Generator 생성

train_generator = ImageDataGenerator(rescale=1/255.0)
valid_generator = ImageDataGenerator(rescale=1/255.0)
```

```python
# 배치 사이즈 만큼 추출
flow_tr_gen = train_generator.flow(X_train, Y_train, batch_size=16, shuffle=True) # 훈련은 랜덤으로 추출
flow_val_gen = valid_generator.flow(X_val, Y_val, batch_size=16, shuffle=False) # 검증 또는 테스트는 순차적으로 추출
```

```python
# ResNet50
# pretrain weight 사용, freeze
from tensorflow.keras.applications import ResNet50

IMAGE_SIZE = 224

base_model = ResNet50(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), weights='imagenet', include_top=False)
bm_output = base_model.output

x = GlobalAveragePooling2D()(bm_output)
x = Dropout(rate=0.1)(x)
x = Dense(200, activation='relu', name='fc1')(x)
x = Dropout(rate=0.1)(x)

#최종 output layer node 수 = class 수
output = Dense(10, activation='softmax', name='output')(x)

model = Model(inputs=base_model.input, outputs=output)
model.summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau , EarlyStopping , ModelCheckpoint , LearningRateScheduler
from tensorflow.keras.optimizers import Adam , RMSprop

model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# 5번 iteration내에 validation loss가 향상되지 않으면 learning rate을 기존 learning rate * 0.2로 줄임.
rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, mode='min', verbose=1)
ely = EarlyStopping(monitor='val_loss', patience=20, mode='min', verbose=1)
```

```python
history = model.fit(flow_tr_gen, epochs=50, validation_data=flow_val_gen, callbacks=[rlr, ely])
```

```python
test_generator = ImageDataGenerator(rescale=1/255.0)
flow_test_gen = test_generator.flow(X_test, Y_test, batch_size=16, shuffle=False)
model.evaluate(flow_test_gen) # loss, accuarcy
```
