```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DM/HW/우리잘했조_동별연령대분포.csv', encoding = 'cp949')
```

```python
# 새로운 데이터프레임 생성
new_df = df.pivot_table(index='동별', columns='연령별', values='2023. 1/4', aggfunc='first')

# 첫 번째 컬럼에 '동별' 정보 추가
new_df.insert(0, '동별', df['동별'].unique())

# 인덱스 이름 변경
new_df.index.name = None
```

```python
new_df.reset_index(inplace = True,drop = True)
```
